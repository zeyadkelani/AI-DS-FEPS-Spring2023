# OLS Regression

Let's say we have the following political data for a certain country:

GDP per capita (in thousands of dollars) <br/>
Political stability index (scale from -2.5 to 2.5)

To analyze the relationship between GDP per capita and the other two variables, and see how they affect GDP per capita.

To do this, we can use the following formula for OLS regression:

y = mx + b

Where:
<br/>

y is the dependent variable (GDP per capita)<br/>
x is the independent variable (political stability index)<br/>
m is the slope of the regression line <br/>
b is the intercept of the regression line <br/>

To find the slope and intercept of the regression line, we can use the following formulas:

m = Σ[(xi - x̄)(yi - ȳ)] / Σ(xi - x̄)²

b = ȳ - m*x̄

Where:

Σ represents the sum of a series of numbers<br/>
xi is the value of the independent variable for observation i<br/>
yi is the value of the dependent variable for observation i<br/>
x̄ is the mean of the independent variable values<br/>
ȳ is the mean of the dependent variable values<br/>

```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
# Input data
gdp = np.array([20, 25, 30, 35, 40, 45, 50, 55, 60, 65])
pol_index = np.array([-2.5, -1.5, -0.5, 0.5, 1.0, 1.5, 2.0, 2.2, 2.4, 2.5])
gdp
```

```python
# Calculate mean of independent variable
x_mean = np.mean(pol_index)

# Calculate mean of dependent variable
y_mean = np.mean(gdp)

# Calculate covariance and variance
covariance = np.sum((pol_index - x_mean) * (gdp - y_mean))
variance = np.sum((pol_index - x_mean) ** 2)

# Calculate slope of regression line
m = covariance / variance

# Calculate intercept of regression line
b = y_mean - m * x_mean

# Print regression equation
print("What's the interpretation of this model?",f"GDP per capita = {m:.2f}*Political stability index + {b:.2f}",sep=os.linesep)
```

### Simplest Demonstration of Machine Learning

```python
# Calculate predicted GDP for a given political stability index
x_test = 1.0  # Change this value to test other values of x
y_pred = m * x_test + b
print(f"Predicted GDP per capita for political stability index of {x_test:.2f}: {y_pred:.2f}")
```

```python
# Plot data and regression line
plt.scatter(pol_index, gdp)
plt.plot(pol_index, m * pol_index + b, color='red')
plt.title('GDP per capita vs Political Stability Index')
plt.xlabel('Political Stability Index')
plt.ylabel('GDP per capita')
plt.show()
```

## What is {m:.2f}?
<br/>
{m:.2f} is a string formatting syntax used to display a floating-point number "m" with 2 decimal places.

```python
m = 3.14159
formatted_m_1 = f"The value of m1 is {m:.1f}"
formatted_m_2 = f"The value of m2 is {m:.2f}"
formatted_m_3 = f"The value of m3 is {m:.3f}"
print(formatted_m_1,formatted_m_2,formatted_m_3, sep=os.linesep )
```
